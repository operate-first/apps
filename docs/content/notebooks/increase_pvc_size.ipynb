# Increase size of a PVC in JupyterHub
This tutorial is used to show how you can increase the size of a PVC (persistent volume claim) in JupyterHub. The first part of this tutorial shows steps from the user's perspective and the second from the operation's perspective.

## User
```{note}
Before creating an issue check if you meet the following requirements:
- [x] User of MOC-Smaug
- [x] Access to JupyterHub on MOC-Smaug
```
As a user, you can increase PVC size by creating [JupyterHub hardware requirements issue](https://github.com/operate-first/support/issues/new?assignees=first-operator&labels=kind%2Fonboarding%2Carea%2FJupyterHub&template=pvc_request.yaml&title=PROJECT%3A+%3Cname%3E) to [operate-first support repository](https://github.com/operate-first/support/issues/86).


## Operation
```{note}
Before you start working on the request check if you meet the following requirements:
- [x] User of MOC-Smaug
- [x] Access to JupyterHub on MOC-Smaug
- [x] User (Requester) requested expansion of a PVC in [operate-first/support/](https://github.com/operate-first/support/)
- [x] Requester is a user of MOC-Smaug
- [x] Requester has access to JupyteHub on MOC-Smaug
```


### Scenarios
When working on the requests, we can encounter two scenarios:
1. Requester uses the default PVC
2. Requester already has a custom PVC defined

### Requester uses the default PVC
<ol>
<li> Get pvc template file from <a href="https://github.com/operate-first/blueprint/blob/main/docs/userstories/jupyterhub/templates/pvc-template.yaml">blueprint/docs/userstories/jupyterhub/templates/pvc-template.yaml </a></li>
<li>Update the highlighted information</li>
  
<pre><code>---
kind: PersistentVolumeClaim
apiVersion: v1
metadata:
  annotations:
    hub.jupyter.org/username: <b>user@redhat.com</b>
  name: <b>jupyterhub-nb-URLENCODED_USERNAME-pvc</b>
  labels:
    app: jupyterhub
    component: singleuser-storage
spec:
  resources:
    requests:
      storage: <b>DESIRED_SIZE</b></code></pre>
      
<li>Encrypt file with SOPS</li>
<li>Add the PVC resource to <a href="https://github.com/operate-first/apps/tree/master/kfdefs/overlays/moc/smaug/opf-jupyterhub/pvcs/"> apps/kfdefs/overlays/moc/smaug/opf-jupyterhub/pvcs/ </a> <br /> following convention <code style="background-color: #f1f1f1;  color: crimson; ">user-USERNAME-pvc.enc.yaml </code>.</li>
<li> List the resource at <a href="https://github.com/operate-first/apps/tree/master/kfdefs/overlays/moc/smaug/opf-jupyterhub/pvcs/secret-generator.yaml">apps/kfdefs/overlays/moc/smaug/opf-jupyterhub/pvcs/secret-generator.yaml</a> </li>
<li>Commit updated files and create PR</li>
<li>When PR is merged and changes are applied by ArgoCD restart requester's JupyterHub server</li>
</ol>


### Requester already has a custom PVC defined
<ol>
<li> Fetch user's PVC resource from  <a href="https://github.com/operate-first/apps/tree/master/kfdefs/overlays/moc/smaug/opf-jupyterhub/pvcs/">apps/kfdefs/overlays/moc/smaug/opf-jupyterhub/pvcs/ </a></li>
<li>Update storage size in file via SOPS</li>
<pre><code>storage: DESIRED_SIZE</code></pre>
<li>Commit updated file and create PR </li>
<li>When PR is merged and changes are applied by ArgoCD restart requester's JupyterHub server</li>
</ol>
